```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = pd.read_csv('17_linux_c.csv', header = 0, low_memory=False)
df = pd.read_csv('17_c.csv', header = 0, low_memory=False)
df_clean = df.copy()
df_clean.info()
```

```python
df_clean = df.drop(columns = "姓名")
X_new= np.array(df_clean)
```

```python
plt.scatter(X_new[:, 0], X_new[:, 1],marker='.')
plt.show()
```

```python
from sklearn.cluster import KMeans

result=KMeans(n_clusters=4,random_state=9).fit_predict(X_new)

# plt.figure(figsize=(5,5))

plt.scatter(X_new[:,0],X_new[:,1],c=result)
```

```python
from sklearn.cluster import DBSCAN

result=DBSCAN(eps=6,min_samples=3).fit_predict(X_new)
plt.scatter(X_new[:,0],X_new[:,1],c=result)
plt.show()
```
